## Objects and Object-Oriented Programming

Object-oriented programming (OOP) is a programming paradigm that is based on the concept of objects. An object is a data structure that has attributes (also called properties) and methods (functions that operate on those properties). OOP is important because it provides a way to organize code and data into reusable, modular pieces that can be used across different parts of a program or in different programs altogether.


here are the components of an object in object-oriented programming:

- Attributes/Properties: These are variables that store the state of an object, such as its size, color, or name.
- Methods/Functions: These are functions that operate on an object's state or behavior, such as changing its size, moving it, or displaying it on screen.
- Constructor: This is a special method that is called when an object is created, and is used to initialize its state and behavior.
- Destructor: This is a special method that is called when an object is destroyed, and is used to clean up any resources that the object is using.
- Class: This is the blueprint or template for creating objects, which defines the attributes, methods, constructor, and destructor that each object of that class will have. 
- Object: This is an instance of a class, created from the blueprint defined by the class, which has its own state and behavior.


OK Now we have a car. Lets add some methods to our Car class so it behaves more like a Car such as start, accelerate. and decelerate

```python
class Car:
    # defining the constructor method that takes in the car's make, model, and year
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.current_speed = 80

    def get_info(self):
        return f'{self.year} {self.make} {self.model}'

    def get_make(self):
        return self.make
    
    # defining a method to increase the car's speed by a given amount
    def accelerate(self, speed_increase):
        self.current_speed += speed_increase
        return self.current_speed

    # defining a method to decrease the car's speed by a given amount
    def decelerate(self, speed_decrease):
        self.current_speed -= speed_decrease
        return self.current_speed
        

    # defining a method to display the car's current speed
    def display_speed(self):
        return self.current_speed
```

In the example above, we have a new Car class that has some components we just learned and a few new ones:

- class (i.e. a specific car object) to modify or retrieve its attributes.
- Attributes: the make, model, year, and current_speed are all attributes of a Car object.
- Constructor method: the __init__ method is a special method that gets called when a new instance of the Car class is created. It sets the initial values for the object's attributes.
- Methods: accelerate, decelerate, and display_speed are methods of the Car class that can be called on an instance of the car

Lets create an instane of our Car and call its methods

```python
# Create an instace of the car 

my_car = Car("Toyota", "Camry", 2021)

print(my_car.make)             # output: "Toyota"
print(my_car.display_speed())  # output: 80

print("Accelerating the car by 10 MPH and the new speed now is:", my_car.accelerate(10))
print("decelerate the car by 20 MPH and the new speed now is:", my_car.decelerate(20))

print(my_car.display_speed()) 
```

### Class Methods vs Functions

In Python, class methods are different from regular functions, although they have some similarities. A class method is a function that belongs to a class rather than an instance of the class. It can be called on the class itself, rather than on an instance of the class, and it can access the class's attributes and methods.

A regular function, on the other hand, is not part of a class and can be called on its own. It cannot access the attributes and methods of a class unless they are passed as arguments.

The key difference between class methods and regular functions is that class methods are bound to the class and have access to its attributes and methods, while regular functions are not bound to any particular class and cannot access its attributes and methods unless they are passed as arguments.

##  Principles of Object-oriented programming (OOP)

Now that we've create a few class and understand instances lest revisit some of the some some of the basic principles of Object-oriented programming (OOP) that are common across many languages, not just Python:

- Encapsulation: This means grouping together data and methods that operate on that data, and hiding the data from the outside world so that it can only be accessed through the methods.
- Abstraction: This means defining an interface that provides a simplified view of a more complex system, so that users can interact with it without needing to understand the underlying details.
- Inheritance: This allows you to define a new class based on an existing class, inheriting its properties and behaviors.
- Polymorphism: This allows you to use a single interface to represent multiple types of objects, so that you can write code that works with different objects in a consistent way.

These principles help to organize code into reusable and modular units, making it easier to maintain and update over time.

The OOP principles are based on the idea of creating objects that encapsulate data and functionality. These objects can interact with each other through well-defined interfaces, which makes it easier to create complex applications.

## inheritance

Inheritance is a concept in object-oriented programming that allows us to create a new class (called a "child" or "sub" class) from an existing class (called a "parent" or "super" class). The child class inherits all of the attributes and behaviors of the parent class, and can also add new attributes and behaviors of its own.

For example, let's say we have a Car class that has attributes like make, model, and year, and behaviors like starting the engine and stopping the engine. We can create a new class called ElectricCar that inherits from the Car class, and adds an attribute for battery size and a behavior for charging the battery.

To define a child class that inherits from a parent class in Python, we use the syntax class ChildClass(ParentClass):. The child class should have its own __init__ method that calls the parent class's __init__ method using the super() function, so that it inherits the parent class's attributes.

Here's an example of how we could define an ElectricCar class that inherits from the Car class:



```python
# Parent Class
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started.")

    def stop_engine(self):
        print("Engine stopped.")

# Child Class
class ElectricCar(Car):
    def __init__(self, make, model, year, battery_size):
        super().__init__(make, model, year)
        
        self.battery_size = battery_size

    def charge_battery(self):
        print("Battery charging.")
        
    def start_engine(self):
        super().start_engine()
        print("Radio turned on.")
```

In the example above, the ElectricCar class inherits from the Car class using the syntax class ElectricCar(Car):. The ElectricCar class also has its own __init__ method that calls the parent class's __init__ method using super(), and adds a new attribute for battery size. It also has a new behavior for charging the battery.

To create an instance of the ElectricCar class, we can use the same syntax as for the Car class, but include the additional argument for battery size:

```python
my_car = Car("Audi", "Q4", 2018)
my_car.start_engine()

my_tesla = ElectricCar("Tesla", "Model S", 2022, "100 kWh")

my_tesla.start_engine()

my_tesla.charge_battery()

my_tesla.stop_engine()
```

In the example above note how we can call methods from both the partner and child classes. Thus the child class - ElectricCar - inherits the attributes of the parent Car class.

n Python, the super() function is used to refer to the parent or superclass of a class. It is used when you want to call a method of a superclass in a subclass. The super() function returns a temporary object of the superclass, which allows you to call its methods.

By calling super(), you can avoid referring to the base class explicitly, which makes your code more maintainable and flexible. This is because if you change the name of the base class, you only need to update it in one place, in the super() call, rather than everywhere the base class is referenced.

For example, let's say you have a Vehicle class that has a method called start_engine(). You then create a Car class that inherits from Vehicle. If you want to call the start_engine() method of the Vehicle class from the Car class, you can use the super() function.

Here's an example:

```python
class Vehicle:
    def start_engine(self):
        print("Engine started.")

class Car(Vehicle):
    def start_engine(self):
        super().start_engine()
        print("Radio turned on.")

my_car = Car()
my_car.start_engine()
```

In this example, the start_engine() method of the Car class calls the start_engine() method of the Vehicle class using super(). This allows the Car class to inherit the behavior of the Vehicle class, while also adding its own behavior.

As you can see, the start_engine() method of the Vehicle class is called first, followed by the print() statement in the start_engine() method of the Car class. This shows how the super() function allows you to use inheritance to reuse code and build on the functionality of existing classes.
