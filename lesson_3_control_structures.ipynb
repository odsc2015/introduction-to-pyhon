# Lesson3 : Control Structures in Python

Control structures are used in programming to control the flow of execution based on certain conditions. In Python, there are three main types of control structures: if, for, and while.

1. <font color=lightblue>if statements</font> are used to execute code if a certain condition is true. They can include else and elif clauses to handle multiple conditions.

2. <font color=lightblue> for loops</font> are used to iterate over a sequence of values, such as a list or a range of numbers. They use a loop variable to take on the value of each element in the sequence in turn.

3. <font color=lighblue>while loops</font> are used to execute a block of code repeatedly as long as a certain condition is true. They can also include a loop variable that is updated on each iteration.

Control structures can be used together and nested to create more complex programs. They are an essential part of programming and allow you to create code that responds dynamically to different situations.

Lets discuss in more detail

## if Statements

The if statement is used to execute code if a certain condition is true. Here's an example:

```python
# if Statements
x = -1

if x > 0:
  print("x is positive")
```

In this example, the **if** statement checks if x is greater than 0. If the condition is true, the code inside the if block (i.e., the print() statement) is executed. If the condition is false, the code inside the if block is skipped.

### else and elif Clauses
In addition to the basic syntax, control structures in Python can also include else and elif clauses to handle multiple conditions. Here's an example:

```python
# if and elif Statements

x = 10

if x < 0:
    print("x is negative")
elif x > 0:
    print("x is positive")
else:
    print("x is zero")
```

#### result
In this example, the if statement checks if x is less than 0, and if so, prints a message. 

If x is not less than 0, the elif statement checks if x is greater than 0, and if so, prints a different message. If neither condition is true, the else block is executed and prints a third message.



## Python Syntax: indentation

Indentation is used to define the block of code that should be executed inside a control structure, such as an if statement, for loop, or while loop. If you don't use indentation, Python will raise an IndentationError and the code will not run.

Here's an example of what happens if you don't use indentation in an if statement:

```python
# Example of what happens if no indentation in an if statement:
x = 5

if x > 0:
print("x is positive")   # Raises IndentationError: expected an indented block
```

#### result

In the example above, the print() statement is not indented after the if statement. When you run this code, Python raises an IndentationError because it expected an indented block of code after the if statement.

Indentation is an important part of Python syntax, and it helps to make the code more readable and understandable. It is recommended to use a consistent and appropriate level of indentation throughout your code. The most commonly used convention is to use four spaces for indentation, which is the standard convention recommended by the Python Software Foundation.

## for Loops

The for loop is used to iterate over a sequence of values, such as a list or a range of numbers. 
Here's an example:

```python
# for loop
fruits = ["apple", "banana", "cherry","Orange"]  # create a simple list

for fruit in fruits:
    print(fruit)
```

#### result

In the example above,  the for loop iterates over the fruits list and prints each fruit on a new line. The loop variable fruit takes on the value of each element in the list in turn.

You can also use the range() function to generate a sequence of numbers to iterate over. For example:

```python
for i in range(5):
    print(i)
```

#### result

In the example above, the for loop iterates over the sequence 0, 1, 2, 3, 4 and prints each number on a new line. The loop variable i takes on the value of each number in the sequence in turn.

###  break statement

```python
fruits = ["apple", "banana", "cherry", "orange", "pear"]

for fruit in fruits:
    if fruit == "apple":
        break
    print(fruit)
    
print("Done!")
```

#### result

The for loop iterates over the fruits list and prints each fruit on a new line, but stops when it encounters the value "cherry". When the break statement is executed, the loop is terminated early and the program continues to execute the code after the loop.

In this output, "cherry","orange", or "pear" is not printed because the break statement causes the loop to exit before it is reached. The code after the loop (i.e., the print("Done") statement) is still executed, because it is outside the loop block.

## while Loops
The while loop is used to execute a block of code repeatedly as long as a certain condition is true. Here's an example:

```python
# while loop
x = 0

while x < 5:
    print(x)
    x += 1
```

#### result

In the example above, the while loop executes the print() statement and increments the value of x by 1 repeatedly as long as x is less than 5. The loop exits when the condition becomes false.

#### While Loop Syntax Explained

 while loop is another type of control structure in Python that allows you to execute a block of code repeatedly as long as a certain condition is true. The basic syntax of a while loop in Python is as follows:

In this syntax, **condition** is the condition that is checked before each iteration of the loop. If the condition is true, the **code block** is executed. The loop continues to iterate as long as the condition is true.

Here are some key features of while loops in Python:

- The **condition** can be any expression that evaluates to a boolean value, such as a comparison or logical operator.
- The **code block** must be indented and can contain any valid Python code, including other control structures such as if statements and nested while loops.
- It is important to make sure that the condition will eventually become false, otherwise the loop will continue to execute indefinitely (known as an "infinite loop").
- You can use the break statement to exit the loop early, or the continue statement to skip to the next iteration of the loop without executing the rest of the code block.

while loops are commonly used in Python to perform operations that require repeated execution, such as searching through a large dataset or implementing a game loop. They are a powerful and flexible tool for controlling the flow of execution in a program.
    
    
### Which One? For or While?
    
As mentioned, the **for** loop is typically used when you want to iterate over a collection of elements such as a list, tuple, or string. The for loop automatically iterates over each element in the collection, and the loop terminates when all elements have been processed.

On the other hand, the **while** loop is used when you want to repeat a set of instructions until a specific condition is met. The loop continues executing as long as the condition specified at the beginning of the loop is True. Once the condition becomes False, the loop will terminate.

So, if you know the number of iterations needed in advance, it's generally better to use a for loop. If you don't know the exact number of iterations needed and need to continue looping until a specific condition is met, then a while loop is the better choice.

## Nesting Control Structures

You can use control structures together in what is called **nesting** to create more complex programs. This simply involves including one control structure inside another. This can be done with any combination of **if** statements, **for** loops, and **while** loops.

Nesting control structures allows you to create more complex programs that respond to multiple conditions and perform multiple tasks. For example, you could use a **for** loop inside an **if** statement to iterate over a sequence of values and perform a certain operation only on values that meet a specific condition 

Here's an example of nesting a for loop inside an if statement, as well as including an elif clause:

```python
# Nesting Control Structures

fruits = ["apple", "banana", "cherry", "orange", "pear","olives"]  # Fun Travia Fact! Olives are an actual fruit@.

for fruit in fruits:
    if "a" in fruit:
        print(fruit)
    elif "o" in fruit:
        print(fruit.upper())
    else:
        print(fruit.capitalize())
```

#### Result

In the above example In this example, the for loop iterates over the fruits list and performs a different operation on each fruit, depending on its contents. 

The if statement checks if the letter "a" is in the fruit string using the in operator, and if so, prints the fruit name. 

The elif clause checks if the letter "o" is in the fruit string, and if so, prints the fruit name in uppercase using the upper() method. The else block is executed if neither condition is true, and capitalizes the first letter of the fruit name using the capitalize() method.

As we can see, the if, elif, and else clauses are nested inside the for loop, creating a more complex program that responds to multiple conditions and performs multiple tasks. The nested control structures make it possible to handle different situations and create more dynamic and responsive programs.
