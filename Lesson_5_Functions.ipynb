## Introduction to Python Functions


In Python, a function is a reusable block of code that performs a specific task. Functions can take inputs, called arguments, and can return a result. Functions can be used to organize code, reduce code duplication, and make code easier to read and understand.

Functions in Python can be thought of as a recipe or set of instructions that performs a specific task. Just like a recipe takes inputs (ingredients) and produces an output (a dish), a function takes inputs (arguments) and produces an output (a return value).

For example, consider a function that adds two numbers together. This function could be thought of as a recipe for adding numbers. The two numbers to be added are the inputs, and the sum of the numbers is the output.

Another analogy for functions is to think of them as machines that take inputs and produce outputs. For example, a vending machine takes money as an input and produces a snack or beverage as an output. Similarly, a function takes arguments as inputs and produces a return value as an output.

```python
# Define a function that calculates the average two numbers

def calculate_average(numbers):
    total = sum(numbers)        # notice the indention
    count = len(numbers)
    average = total / count
    return average             # retun the calculated average
```

### More Function Arguments

As mentioned, functions can take zero or more arguments, which are inputs to the function that affect its behavior. Arguments can be specified as positional arguments (in the order they are defined) or as keyword arguments (specified by name).

For example, here's a function that takes two positional arguments and returns their product:

```python
# Define a function that multiplies two numbers

def multiply_numbers(x, y):
    print("x argument value is ", x)
    print("y argument value is ", y)
    return x * y

# Call the multiply_numbers function with positional arguments

result1 = multiply_numbers(3, 5)

print(result1)  # Output: 15

# Call the multiply_numbers function with keyword arguments. Note the order is switched.

result2 = multiply_numbers(y=4, x=6)

print(result2)  # Output: 24
```

In this example, the multiply_numbers function takes two arguments, x and y, and returns their product. We can call the function with two positional arguments, or with two keyword arguments.

### What to Watch For

There are a few common things that you should watch out for when defining functions in Python:

- **Syntax errors:** When defining a function, it's important to make sure that the syntax is correct, including the placement of parentheses, commas, and colons.

- **Naming conventions:** Function names should be descriptive and follow the same naming conventions as variable names, using lowercase letters and underscores to separate words.

- **Arguments:** If a function takes arguments, it's important to make sure that the number and type of arguments are specified correctly in the function header.

- **Indentation:** Python uses indentation to indicate the scope of code blocks, so it's important to make sure that the code inside the function is indented correctly.

- **Return statement:** If a function is meant to return a value, make sure that the return statement is included and that the value being returned is the correct data type.

- **Function calls:** Make sure that the function is called correctly, including the correct number and type of arguments, and that the returned value is used appropriately in the calling code.

Good function names:

- calculate_average
- convert_to_celsius
- display_message
- load_data

These names are good because they are descriptive and clearly indicate what the function does. They are also written in lowercase letters with underscores separating words, which is the standard naming convention for functions in Python.

Bad function names:

- abc
- function1
- my_func
- do_stuff

These names are bad because they are not descriptive and do not indicate what the function does. They are also not written using the standard naming convention for functions in Python. It's important to choose a descriptive and meaningful name for your function so that other programmers can understand what your code does.


Overall, by paying attention to these common issues, beginners can avoid common mistakes when defining functions in Python.

This is a follow along exercise. Lets write a program that calculates the average temperature of a list of daily temperatures.

### Step 1: Defining Variables
First, let's define a list of daily temperatures:

```python
# Define a list of tempratures. 

temperatures = [72, 75, 68, 70, 73, 72, 75]
```

We'll use this list to calculate the average temperature for the week.

### Step 2: Using Data Structures
Next, we'll use a function to calculate the average temperature. Here's the function definition:

```python
# Define a function to calculate the average temperature. 

def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    average = total / count
    return average
```

This function takes a list of numbers as an argument and returns the calculated average value.

### Step 3: Using Control Structures
To calculate the average temperature for the week, we'll call the calculate_average function with the temperatures list as an argument:

```python
# call the calculate_average function

average_temperature = calculate_average(temperatures)
```

Next, we'll use a conditional statement to display a message to the user depending on whether the average temperature is above or below a certain threshold:

```python
if average_temperature > 72:
    print("It's been a warm week!")
else:
    print("It's been a cooler week.")
```

This conditional statement checks whether the average_temperature variable is greater than 72 degrees Fahrenheit. If it is, the program will display a message indicating that it's been a warm week. If it's not, the program will display a message indicating that it's been a cooler week.

## Step 4: Putting it All Together
Here's the complete program:

```python
# Define a list of daily temperatures
temperatures = [72, 75, 68, 70, 73, 72, 75]

# Define a function to calculate the average temperature
def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    average = total / count
    return average

# Calculate the average temperature for the week
average_temperature = calculate_average(temperatures)

# Display a message to the user based on the average temperature
if average_temperature > 72:
    print("It's been a warm week!")
else:
    print("It's been a cooler week.")
```
